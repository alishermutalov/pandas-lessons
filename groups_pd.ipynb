```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data.csv', index_col=0)
df.head()
```

```python
df.price.describe()
```

```python
bins = [0, 10000, 20000, 50000]
df['price_bin'] = pd.cut(df.price, bins)
df.head()
```

```python
names = ["arzon", "o'rta", "qimmat"]
df['price_bin_name'] = pd.cut(df.price, bins, labels=names)
df['price_bin'] = pd.cut(df.price, bins)
df.head()
```

```python
range = pd.cut(df.price, 3)
range.value_counts()
```

```python
range = pd.qcut(df.price, 3)
range.value_counts()
#pd.cut va pd.qcut funksyalarining farqi: cut -> guruhlarga ajratadi (ammo har bir guruhda teng balans bo'lmaydi)
#qcut -> guruhlarga ajratadi (ammo har bir guruhda teng balans bo'ladi)
#misol:
#pd.cut(df.price, 3)
#   taxminan : group1 = 44, group2 = 9, group3 = 5
#pd.qcut(df.price, 3)
#   taxminan : group1 = 20, group2 = 19, group3 = 19
```
