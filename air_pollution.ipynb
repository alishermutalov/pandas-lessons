```python
df = pd.read_csv('/content/Practicum_Datasets/world_cities_air_water_quality.csv')
df.head()
```

```python
df.columns = df.columns.str.replace('"', '')
```

```python
df
```

```python
df.columns = df.columns.str.strip()
df[['Region', 'Country']]=df[['Region', 'Country']].map(lambda x: x.replace('"', ''))
```

```python
df[['Region', 'Country']]=df[['Region', 'Country']].map(lambda x: x.strip())
```

```python
df
```

```python
# df['Country']
df1 = pd.pivot_table(df, index=['Country','City'], aggfunc='first')
df1
```

```python
df2 = df1.copy()
```

```python
df2
```

```python
df2['AirPollution'] = 100-df2['AirQuality']
df2['WaterQuality'] = 100-df2['WaterPollution']
df2.drop('Region', axis=1, inplace=True)
df2
```

```python
df3 = pd.DataFrame(df2.values, index=df2.index, columns=[['AIR', 'AIR', 'WATER', 'WATER'], ['AirQuality', 'AirPollution', 'WaterPollution', 'WaterQuality']])
df3
```

```python
df3.columns.names = ['Object', 'Info']
df3
```

```python
df3.swaplevel('Country','City')
```

```python
df3.sort_index(level='Country', ascending=False) #level=0 qilish orqali ham biz df ni o'zgartirishimiz mumkin
```

```python
#Ko'p darajali indekslar bilan bir vaqtda bir nechta darajalar bo'yicha
#tartiblash uchun level ga bir nechta darajalarni ko'rsatishingiz mumkin:
df3.sort_index(level=['Country', 'City'], ascending=[False, True])
```
