```python
transport_price = pd.DataFrame({
    'Transport':['poyezd', 'tramvay','kema'],
    'Narxi':[12000000, 800000, 2500000]
})
transport_price
```

```python
transport_speed = pd.DataFrame({
    'Transport':['poyezd', 'tramvay','samalyot'],
    'Tezligi':[350, 60, 800]
})
transport_speed
```

```python
pd.merge(transport_price, transport_speed) #inner join
```

```python
pd.merge(transport_price, transport_speed, how='outer') #outer join
```

```python
pd.merge(transport_price, transport_speed, how='left') #left join
```

```python
pd.merge(transport_price, transport_speed, how='right') #right join
```

```python
transport_price = pd.DataFrame({
    'Transport_turi':['poyezd', 'tramvay','kema'],
    'Narxi':[12000000, 800000, 2500000]
})
transport_price
```

```python
transport_speed = pd.DataFrame({
    'Transport':['poyezd', 'tramvay','samalyot'],
    'Tezligi':[350, 60, 800]
})
transport_speed
```

```python
pd.merge(transport_price,transport_speed, left_on='Transport_turi', right_on='Transport')
```

```python
df = pd.merge(transport_price,transport_speed, left_on='Transport_turi', right_on='Transport')
df.drop(columns='Transport', inplace=True)
df
```
